```python
from transformers import pipeline
from bs4 import BeautifulSoup
import requests
```

```python
summarizer=pipeline("summarization")
```

```python
url="https://hackernoon.com/the-epitome-of-loyalty-marketing-samurais-and-the-bushido-code"
r=requests.get(url)
```

```python
r.text
```

```python
soup=BeautifulSoup(r.text,"html.parser")
res=soup.find_all(["h1","p"])
res
```

```python
text=[re.text for re in res]
arti=" ".join(text)
```

```python
arti=arti.replace(".",".<eos>")
arti=arti.replace("!","!<eos>")
arti=arti.replace("?","?<eos>")
sente=arti.split("<eos>")
```

```python
sente[2]
```

```python
max_chunk=400
current_chunk=0
chunks=[]
for sent in sente:
  if len(chunks)== current_chunk+1:
    if len(chunks[current_chunk])+len(sent.split(" ")) <= max_chunk:
      chunks[current_chunk].extend(sent.split(" "))
    else :
      current_chunk+=1
      chunks.append(sent.split(" "))
  else:
    print(current_chunk)
    chunks.append(sent.split(" "))
```

```python
chunks[0]
```

```python
for chunk_id in range(len(chunks)):
  chunks[chunk_id]=" ".join(chunks[chunk_id])
```

```python
chunks[0]
```

```python
len(chunks)
```

```python
ress=summarizer(chunks,max_length=120,min_length=30,do_sample=False)
```

```python
ress
```

```python
text=" ".join([summ["summary_text"] for summ in ress])
```

```python
with open("blog_summary.txt","w") as f:
  f.write(text)
```
